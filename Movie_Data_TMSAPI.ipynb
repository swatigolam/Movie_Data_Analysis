```python
#create connection

conn = sqlite3.connect('Movies_Data_Assignment.db')
c = conn.cursor()
```

```python
engine = create_engine('sqlite:///Movies_Data_Assignment.db', echo = True)
```

```python
meta = MetaData()
```

```python
#create tables

Movies_Data_Theaters = Table(
   'Movies_Data_Theaters', meta, 
   Column('id', String), 
   Column('Release_year', Integer), 
   Column('Title', String),
   Column('Generes', String),
   Column('Description', String),
   Column('Theatre_Channel', String),
)
meta.create_all(engine)
```

```python
#create tables

Movies_Data_TV = Table(
   'Movies_Data_TV', meta, 
   Column('id', String,), 
   Column('Release_year', Integer), 
   Column('Title', String),
   Column('Generes', String),
   Column('Description', String),
   Column('Theatre_Channel', String),
)
meta.create_all(engine)
```
